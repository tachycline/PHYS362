## Problem statement
In this problem we explore some of the more useful theorems (stated without proof) involving Hermite polynomials.

(a) The Rodrigues formula says that 
 $$ H_n(\xi) = (-1)^ne^{\xi^2}(d/d\xi)^ne^{(-\xi^2)}$$
 Use it to derive $H_3$ and $H_4$

(b) The following recursion relation gives you $H_{n+1}$ in terms of the two preceding Hermite polynomials:
 $$ H_{n+1})(\xi) = 2\xi H_n(\xi) -2nH_{n-1}(\xi) $$
 Use it, together with your answer in (a) to obtain $H_5$ and $H_6$

(c) If you differentiate an nth-order polynomial, you get a polynomial of order (n-1). For the Hermite polynomials, in fact,
 $$ dH_n/d\xi = 2nH_{n-1}(\xi)$$
 Check this, by differentiating $H_5 and H_6$

(d) $H_n(\xi)$ is the nth z-derivative, at z = 0, of the generating function exp($-z^2+2z\xi$) or, to put it another way, it is the coefficient of $z^n/n!$ In the Taylor series expansion for this function:
 $$ e^{-z^2+2z\xi} = \sum_{n = 0}^{\infty}z^nH_n(\xi)/n!$$
 Use this to rederive $H_0,H_1, and H_2$

## Solution


```python
import sympy as sp
sp.init_printing()

xi = sp.symbols('xi', real=True, positive=True)
n = sp.symbols('n', real=True, integer=True)
z = sp.symbols('z')
```

```python
# plug in n=3, n=4 to the Rodrigues Formula
n = 3
H_n = (-1)**n*sp.exp(xi**2)*sp.diff(sp.exp(-xi**2), xi, n)
sp.simplify(H_n)
```

So $H_3 = 8\xi^3 - 12\xi$

```python
n = 4 
H_n = (-1)**n*sp.exp(xi**2)*sp.diff(sp.exp(-xi**2), xi, n)
sp.simplify(H_n)
```

So $ H_4 = 16\xi^4 - 48\xi^2 + 12 $

```python
# use recursion formular to calculate H5
H_4 = 16*xi**4 - 48*xi**2 +12
H_3 = 8*xi**3 - 12*xi
H_5 = 2*xi*H_4 - 2*4*H_3
sp.simplify(H_5)
```

Here we got $H_5$

```python
H_6 = 2*xi*H_5 - 2*5*H_4
sp.simplify(H_6)
```

Here we got $H_6$

```python
#Check hermite polynomials by differentiation
n = 6
sp.simplify(sp.diff(H_6, xi)/(2*n))
```

Which is the same with $H_5$

```python
n = 5 
sp.simplify(sp.diff(H_5)/(2*n))
```

Which is equal to $H_4$

```python
z = sp.symbols('z')
n = 0
H = sp.diff(sp.exp(-z**2+2*z*xi), z, n)
H = H.subs(z,0)
H
```

Here we got $H_0$

```python
z = sp.symbols('z')
n = 1
H = sp.diff(sp.exp(-z**2+2*z*xi), z, n)
H = H.subs(z,0)
H
```

Here we got $H_1$

```python
z = sp.symbols('z')
n = 2
H = sp.diff(sp.exp(-z**2+2*z*xi), z, n)
H = H.subs(z,0)
H
```

Here we got $H_2$
