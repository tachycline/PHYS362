# Problem 1.9

## A particle of mass m is in the state

##  $$ \Psi(x,t)=Ae^{-a[(mx^2/\hbar)+it]}$$

## where A and a are positive real constants.
## a) Find A.
## b) For what potential energy function V(x) does $\Psi$ satisfy the Schrodinger equation?
## c) Calculate the expecation values of x, $x^2$, p, and $p^2$.
## d) Find $\sigma_x$ and $\sigma_p$. Is their product consistent with the uncertainty principle?



----

## Solution
### A) Finding A

```python
import sympy as sp
from sympy.physics.quantum.constants import hbar

#pretty output
sp.init_printing()

#Setting up symbols
A, a, m, = sp.symbols('A a m', positive=True, real=True)
x, t = sp.symbols('x t', real=True)

#Defining Psi
Psi = A*sp.exp(-a*((m*(x**2)/hbar)+sp.I*t))
Psi
```

```python
#In order for wave function to be normalized, must be equal to 1
normalization_condition = sp.solveset(sp.Eq(1, sp.integrate(Psi * sp.conjugate(Psi), (x, -sp.oo, sp.oo))),A)

#Solving for A
normalization_condition                                    
```

```python
Psi = Psi.subs(A, next(iter(normalization_condition)))
Psi
```

### B) For what V(x) does $\Psi$ satisfy the Schrodinger equation? 

```python
dPsidt = sp.diff(Psi, t)
dPsidt
```

```python
dsqPsidxsq = sp.diff(Psi, x, 2)
dsqPsidxsq
```

```python
V = sp.symbols("V", real=True)
schro = sp.Eq(sp.I*hbar*dPsidt, -hbar**2/(2*m)*dsqPsidxsq + V*Psi)
schro
```

```python
sp.solveset(schro, V)
```

```python
sp.Eq(V, next(iter(sp.solveset(schro, V))).simplify())
```

## c) Find the expectation values for x, x^2, p, p^2

```python
expectx = sp.integrate(sp.conjugate(Psi)*x*Psi, (x, -sp.oo, sp.oo))
expectxsq = sp.integrate(sp.conjugate(Psi)*x**2*Psi, (x, -sp.oo, sp.oo))

expectp = -hbar*sp.I*sp.integrate(sp.conjugate(Psi)*sp.diff(Psi, x), (x, -sp.oo, sp.oo))
expectpsq = -hbar**2*sp.integrate(sp.conjugate(Psi)*sp.diff(Psi, x, 2), (x, -sp.oo, sp.oo))

expectx
```

## d) Find $\sigma_x$ and $\sigma_p$. Is their product consistent with the uncertainty principle?

```python
sigmax = sp.sqrt(expectxsq - expectx**2)
sigmap = sp.sqrt(expectpsq - expectp**2)
sigmax
```

```python
sigmap
```

```python
sigmax*sigmap
```
